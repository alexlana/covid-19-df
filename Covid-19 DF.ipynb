```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data.tsv', sep='\t')
```

```python
df
```

```python
plt.figure(figsize=(16, 6))
plt.xticks(rotation=70)
plt.tight_layout()

sns.set(style="whitegrid")
ax = sns.barplot(x="date", y="total_infected", data=df,  color="blue", saturation=.5)

```

```python
plt.figure(figsize=(16, 6))
plt.xticks(rotation=70)
plt.tight_layout()

ax = sns.lineplot(x="date", y="total_infected", data=df, sort=False)
```

```python
plt.figure(figsize=(16, 6))
plt.xticks(rotation=70)
plt.tight_layout()

ax = sns.lineplot(x="date", y="total_infected", data=df, sort=False)
ax.set_yscale('log')
```
